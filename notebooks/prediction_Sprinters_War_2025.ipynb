# üèÅ Sprinters' War ‚Äì Predictive Modeling for a Fictional Sprint League

**Project Overview**  
This notebook is part of a fun, exploratory data science project built around *Sprinters' War* ‚Äî a **fictitious 100m sprint competition** featuring the world‚Äôs top male sprinters in a head-to-head knockout format. Inspired by the showmanship of events like the NBA All-Star Game, *Sprinters' War* blends elite performance with entertainment, stats immersion, and simulation storytelling.

**Context**  
The event does **not actually exist** (yet), but serves as a creative playground to test predictive features, simulate outcomes, and tell performance-based stories using real-world sprint data (e.g., from World Athletics). The 2025 edition is imagined to take place in Paris with a tournament bracket including 8 quarterfinal duels, progressing to semifinals and a final.

‚ö†Ô∏è *Disclaimer*: This is not an optimized or production-grade modeling pipeline. It's a sandbox project meant for experimentation, fun, and sprint performance geekery.

**Objectives of the Notebook**:
- Simulate the full tournament by predicting race outcomes for each round
- Use real historical data to create synthetic matchups
- Predict 100m times based on contextual features
- Dynamically update predictions as the tournament unfolds

**Key Features**:

The model uses a comprehensive set of contextual and performance-related features to estimate sprint times. These include:

- `Name` and `Country`: athlete identity and nationality  
- `Date`: date of the race (for temporal consistency)  
- `Mark_clean`: clean version of the time recorded (used as target during training)  
- `Age`: athlete's age on race day  
- `Wind_float`: wind assistance or resistance in m/s  
- `TimeGap`: gap between the predicted performance level of the two athletes  
- `HomeRace`: binary flag indicating whether the athlete is competing in their home country  
- `RoundType`: type of race round (e.g., heats, semifinals, final, duel)  
- `IsMajorEvent`: whether the competition is a major championship or not  
- `Nation`: used to encode nationality-based effects  
- `PB_so_far`: personal best of the athlete at that point in time  
- `SB_so_far`: seasonal best of the athlete at that point in time  
- `avg_last_3`: average time of the last 3 races (recent form indicator)  
- `race_season_rank`: rank of the current race within the athlete's season (e.g., 1st, 2nd...)  
- `win_ratio_against_*`: a set of binary features indicating past win ratios against each possible opponent in the bracket. These serve as a proxy for rivalry effects and psychological advantage.

This feature set is designed to balance **historical performance**, **season context**, **external conditions**, and **duel dynamics**, providing a rich input space for the quantile regression model.

**Modeling Approach Used**:
- ‚è± **Quantile Regression** is the sole modeling method applied in this notebook. It allows us to estimate the distribution of possible race times (e.g., median or lower/higher percentiles), helping simulate outcomes with a notion of uncertainty without relying on probabilistic models.

**Why this matters** (even if it's fictional üòâ):  
This kind of sandbox allows for testing how predictive features perform in sprint events, understanding race dynamics, and iterating creatively on sports analytics in a domain (track & field) that is often underexplored from a data science perspective.

```python
# Conversion s√©curis√©e des colonnes
df["DOB"] = pd.to_datetime(df["DOB"], errors="coerce", format="%d %b %Y")
df["Date"] = pd.to_datetime(df["Date"], errors="coerce")

# Calcul de l'√¢ge au moment de la course
df["Age"] = ((df["Date"] - df["DOB"]).dt.days / 365.25).round(2)

# Fonction pour extraire le statut de course (F = Finished, dns, dnf, dq)
def extract_status(mark):
    mark = str(mark).strip().upper()
    if mark == "DNS":
        return "dns"
    elif mark == "DNF":
        return "dnf"
    elif mark == "DQ":
        return "dq"
    else:
        return "F"  # Finished

# Appliquer le statut
df["race_status"] = df["Mark"].apply(extract_status)

# Identifier si la performance est homologu√©e (absence d'ast√©risque)
df["is_legal"] = df["Mark"].str.contains(r"\*").apply(lambda x: 0 if x else 1)

# Supprimer les ast√©risques
df["Mark_sanitized"] = df["Mark"].str.replace("*", "", regex=False)

# Convertir en float (chrono propre), NaN pour DNS/DNF/DQ
df["Mark_clean"] = pd.to_numeric(df["Mark_sanitized"], errors='coerce')

# Supprimer la colonne interm√©diaire
df.drop(columns=["Mark_sanitized"], inplace=True)

# Conversion du vent en float
df["Wind_float"] = pd.to_numeric(df["Wind"].str.replace("+", "", regex=False), errors="coerce")

# Nettoyage de la colonne Place
df["Place"] = pd.to_numeric(df["Place"], errors="coerce")

# D√©tection des courses √† domicile
df["HomeRace"] = df["Country"] == df["Nation"]

df.head()
```

```python
def get_round(race_code):
    if isinstance(race_code, str):
        race_code = race_code.upper()
        if race_code.startswith("H"):
            return "Heats"
        elif race_code.startswith("SF"):
            return "Semi-Final"
        elif race_code.startswith("F"):
            return "Final"
        elif race_code.startswith("QF"):
            return "Quarter-Final"
        elif race_code.startswith("PR"):
            return "Preliminary Round"
        elif race_code.startswith("ER"):
            return "Extra Race"
        else:
            return "Other"
    return "Unknown"

df["RoundType"] = df["Race"].apply(get_round)

df.head()
```

```python
# D√©tection des √©v√©nements majeurs (JO, Mondiaux, Diamond League)
def is_major_event(competition_name):
    if isinstance(competition_name, str):
        major_keywords = ["Olympic Games", "World Championships", "Diamond Discipline"]
        return any(keyword in competition_name for keyword in major_keywords)
    return False

df["IsMajorEvent"] = df["Competition"].apply(is_major_event)

df.sort_values(by=["Name", "Date"], ascending=[True, True]).head()
```

```python
# 1. D√©finir la priorit√© des rounds
round_priority = {
    "Preliminary Round": 1,
    "Heats": 2,
    "Quarter-Final": 3,
    "Semi-Final": 4,
    "Final": 5,
    "Extra Race": 6,
    "Unknown": 99
}
df["RoundPriority"] = df["RoundType"].map(round_priority)
```

```python
# Tri initial pour TimeGap ‚Äî ne pas toucher
df = df.sort_values(by=["Name", "Date", "RoundPriority"], ascending=[True, False, False])

# TimeGap
df["PrevDate"] = df.groupby("Name")["Date"].shift(-1)
df["PrevYear"] = df.groupby("Name")["Year"].shift(-1)
df["TimeGap"] = (df["Date"] - df["PrevDate"]).dt.days
df.loc[df["Year"] != df["PrevYear"], "TimeGap"] = np.nan

# Identifiant temporaire
df["temp_id"] = df.index

# üß† Tri chronologique r√©el pour course dans saison : date ‚Üë, round ‚Üì
df_rank = (
    df.sort_values(by=["Name", "Year", "Date", "RoundPriority"], ascending=[True, True, True, False])
      .copy()
)
df_rank["race_season_rank"] = df_rank.groupby(["Name", "Year"]).cumcount() + 1

# Projection dans le df initial
df = df.merge(df_rank[["temp_id", "race_season_rank"]], on="temp_id", how="left").drop(columns="temp_id")
```

```python
# Correction des cas incoh√©rents √† la marge
mask_first_wrong = df["TimeGap"].isna() & (df["race_season_rank"] == 2)
mask_second_wrong = (df["TimeGap"] == 0.0) & (df["race_season_rank"] == 1)

df.loc[mask_first_wrong, "race_season_rank"] = 1
df.loc[mask_second_wrong, "race_season_rank"] = 2
```

```python
# Initialisation des nouvelles colonnes
df_perf['SB_so_far'] = None
df_perf['PB_so_far'] = None
df_perf['avg_last_3'] = None
df_perf['is_first_race_season'] = False

# Traitement par athl√®te
for name, group in df_perf.groupby('Name'):
    group = group.sort_values('Date')
    all_marks = []
    for year, season_df in group.groupby(group['Date'].dt.year):
        season_df = season_df.sort_values('Date')
        marks_season = []
        for i, (idx, row) in enumerate(season_df.iterrows()):
            if i == 0:
                df_perf.loc[idx, 'is_first_race_season'] = True
            else:
                last_3 = marks_season[-3:]
                df_perf.loc[idx, 'avg_last_3'] = sum(last_3) / len(last_3) if last_3 else None
                df_perf.loc[idx, 'SB_so_far'] = min(marks_season) if marks_season else None
            df_perf.loc[idx, 'PB_so_far'] = min(all_marks) if all_marks else None
            if pd.notna(row['Mark_clean']):
                marks_season.append(row['Mark_clean'])
                all_marks.append(row['Mark_clean'])
                
df_perf['RaceID'] = df_perf['Date'].astype(str) + "_" + df_perf['Competition'] + "_" + df_perf['Race']

df_perf.head(20)
```

```python
df_filtered = df_perf[['Name', 'Nation', 'Date', 'Competition', 'Race', 'Mark_clean', 'RaceID']].copy()

# Dictionnaires pour compter les duels et les victoires
victory_counts = defaultdict(lambda: defaultdict(int))
duel_counts = defaultdict(lambda: defaultdict(int))

# Parcours de chaque course pour extraire les duels
for race_id, group in df_filtered.groupby('RaceID'):
    sorted_group = group.sort_values('Mark_clean')
    athletes = list(sorted_group['Name'])
    
    for i in range(len(athletes)):
        winner = athletes[i]
        for j in range(i + 1, len(athletes)):
            loser = athletes[j]
            victory_counts[winner][loser] += 1
            duel_counts[winner][loser] += 1
            duel_counts[loser][winner] += 1  # car B a aussi couru contre A

# Construction du DataFrame final avec win_ratios
rows = []
for athlete in df['Name'].unique():
    row = {'Name': athlete}
    for opponent in df['Name'].unique():
        if athlete == opponent:
            continue
        wins = victory_counts[athlete][opponent]
        total_duels = duel_counts[athlete][opponent]
        if total_duels > 0:
            row[f'win_ratio_against_{opponent}'] = wins / total_duels
        else:
            row[f'win_ratio_against_{opponent}'] = np.nan
    rows.append(row)

df_ratios = pd.DataFrame(rows)

# Ajout de la Nation
nation_map = df[['Name', 'Nation']].drop_duplicates().set_index('Name')['Nation']
df_ratios['Nation'] = df_ratios['Name'].map(nation_map)

# R√©organisation des colonnes
cols = ['Name', 'Nation'] + [col for col in df_ratios.columns if col.startswith('win_ratio_against_')]
df_ratios = df_ratios[cols]

df_full = df_perf.merge(df_ratios, how='left', on='Name')

df_full = df_full.drop(columns='Nation_y')  # on supprime la colonne inutile
df_full = df_full.rename(columns={'Nation_x': 'Nation'})

# S√©lection des colonnes utiles pour le mod√®le
features_finales = ['Name','Country', 'Date', 'Mark_clean', 'Age', 'Wind_float', 'TimeGap', 
                    'HomeRace', 'RoundType', 'IsMajorEvent', 'Nation', 'PB_so_far', 'SB_so_far', 
                    'avg_last_3', 'race_season_rank'] + \
                   [col for col in df_ratios.columns if col.startswith('win_ratio_against_')]

df_Xit = df_full[features_finales]
```

# Entrainement du mod√®le

```python
# S√©paration features / target
X = df_Xit.drop(columns=['Mark_clean', 'Name', 'Date','Country'])
y = df_Xit['Mark_clean']
meta = df_Xit[['Name', 'Date']]

# Encodage des variables cat√©gorielles
X_encoded = pd.get_dummies(X)

# Split train/test avec meta
X_train, X_test, y_train, y_test, meta_train, meta_test = train_test_split(
    X_encoded, y, meta, test_size=0.2, random_state=42
)

# Fonction d'entra√Ænement
def train_quantile_model(alpha):
    model = HistGradientBoostingRegressor(loss="quantile", quantile=alpha,
                                          learning_rate=0.05, max_iter=300,
                                          max_leaf_nodes=31, min_samples_leaf=5,
                                          random_state=42)
    model.fit(X_train, y_train)
    return model

# Entra√Ænement des 3 mod√®les
model_q05 = train_quantile_model(0.05)
model_q50 = train_quantile_model(0.50)
model_q95 = train_quantile_model(0.95)

# Pr√©dictions
pred_q05 = model_q05.predict(X_test)
pred_q50 = model_q50.predict(X_test)
pred_q95 = model_q95.predict(X_test)

# Cr√©ation du tableau des intervalles + vraies valeurs + contexte
interval_df = pd.DataFrame({
    'Name': meta_test.reset_index(drop=True)['Name'],
    'Date': meta_test.reset_index(drop=True)['Date'],
    'q05': pred_q05,
    'q50': pred_q50,
    'q95': pred_q95,
    'true': y_test.reset_index(drop=True)
})

# Aper√ßu
print(interval_df.head())

# 1. √âvaluation q50
mae = mean_absolute_error(interval_df['true'], interval_df['q50'])
rmse = np.sqrt(mean_squared_error(interval_df['true'], interval_df['q50']))
r2 = r2_score(interval_df['true'], interval_df['q50'])

print("‚úÖ Performance sur la m√©diane (q50) :")
print(f"MAE  = {mae:.4f}")
print(f"RMSE = {rmse:.4f}")
print(f"R¬≤   = {r2:.4f}")

# 2. √âvaluation des intervalles
inside_interval = (interval_df['true'] >= interval_df['q05']) & (interval_df['true'] <= interval_df['q95'])
coverage = inside_interval.mean()
avg_width = (interval_df['q95'] - interval_df['q05']).mean()

print("\n‚úÖ Performance sur l'intervalle [q05, q95] :")
print(f"Taux de couverture (true ‚àà [q05, q95]) = {coverage:.2%}")
print(f"Largeur moyenne de l'intervalle       = {avg_width:.4f}")
```

## üß™ Simulation Workflow ‚Äì How a Sprinters' War Edition is Simulated

Each edition of *Sprinters' War* simulates a **matchday in a closed sprint league**, where every athlete competes in exactly one duel per edition, followed by a second phase with either a second duel or a solo run.

---

### üåç Structure of an Edition

There are **8 editions in the season**, each hosted in a different iconic city:

**Paris ‚Äì New York ‚Äì Kingston ‚Äì Tokyo ‚Äì London ‚Äì Nairobi ‚Äì Doha ‚Äì Zurich**

Each edition follows this structure:

#### üîπ First Duel Series (S√©rie 1)
- **8 duels** are defined randomly (via pre-generated match program)
- All 16 athletes compete once
- Each race is simulated using contextual features and a quantile regression model

#### üî∏ Second Duel Series (S√©rie 2)
- **7 duels** are randomly defined using the 14 athletes not matched previously
- The **2 remaining athletes** (not drawn in any duel) run a **solo race**  
- The solo race counts toward the performance points but awards no duel victory points

---

### üèÖ Points System

Each edition awards **two types of points**:

#### ü•á Duel Points
- Victory in a duel ‚Üí **3 points**
- Defeat in a duel ‚Üí **0 points**
- No duel in solo race ‚Üí **0 points**

#### ‚è± Performance Bonus (based on times from both series)
- Fastest time of the edition ‚Üí **3 points**
- 2nd fastest time ‚Üí **2 points**
- 3rd fastest time ‚Üí **1 point**

Solo runs are **eligible for performance points**, making them strategically important despite having no opponent.

---

### ‚öôÔ∏è Feature Setup & Simulation Details

Before each edition, the simulation engine sets up the following:

1. **Edition Parameters**
   - Date of the edition (`event_date`)
   - Random wind (`Wind_float`) uniformly drawn between -2 and +2 m/s

2. **Best Performance So Far**
   - `SB_so_far`: best simulated time (q50) in all previous races
   - `PB_so_far`: updated if SB beats prior PB

3. **Athlete Profile Updates**
   - `Age`: recalculated based on edition date
   - `Date`: assigned to edition date
   - `TimeGap`: 0 for 1st duel of the night, 15 for 2nd duel or solo run
   - `race_season_rank`: incremented at each new race
   - `IsMajorEvent`: always `True`
   - `RoundType`: always set to `"Final"` for stylistic consistency
   - `HomeRace`: `True` if the athlete‚Äôs country matches the host nation

4. **Recent Form**
   - `avg_last_3`: mean of last 3 `q50` times (simulated), calculated from historical results

5. **Opponent-Specific Features**
   - `win_ratio_against_*`: updated after each edition using head-to-head duel outcomes

---

This process ensures each race is contextualized within the broader league narrative. The use of **quantile regression (q=0.5)** ensures realistic yet flexible race time simulation, and the structure blends duel tension with individual time trial stakes ‚Äî in line with the show-first philosophy of *Sprinters' War*.


```python
def add_ranking_movement(df_current, df_previous):
    # Cr√©er rangs
    df_current = df_current.copy()
    df_current["Rank"] = df_current["TotalPoints"].rank(ascending=False, method="min").astype(int)
    df_previous = df_previous.copy()
    df_previous["PrevRank"] = df_previous["TotalPoints"].rank(ascending=False, method="min").astype(int)

    # Fusion des deux classements
    df_merged = df_current.merge(df_previous[["Name", "PrevRank"]], on="Name", how="left")

    # D√©terminer le mouvement
    def get_arrow(row):
        if pd.isna(row["PrevRank"]):
            return "üÜï"  # nouveau classement
        delta = row["PrevRank"] - row["Rank"]
        if delta > 0:
            return f"üîº +{delta} üü¢"
        elif delta < 0:
            return f"üîΩ {abs(delta)} üî¥"
        else:
            return "‚ûñ 0 ‚ö™"

    df_merged["Mouvement"] = df_merged.apply(get_arrow, axis=1)

    # Trier par classement final
    df_final = df_merged.sort_values("Rank").reset_index(drop=True)
    df_final["üèÖ Rang"] = df_final.index + 1
    df_final = df_final[["üèÖ Rang", "Name", "Mouvement", "NumRaces", "TotalPoints"]]

    return df_final
```

```python
# === 1. Infos de base ===
event_date = pd.to_datetime("2025-04-01")
athletes = [
    "Kendal WILLIAMS", "Kishane THOMPSON",
    "Christian MILLER", "Ferdinand OMANYALA",
    "Emmanuel MATADI", "Noah LYLES",
    "Shainer REGINFO MONTOYA", "Fred KERLEY",
    "Ackeem BLAKE", "Oblique SEVILLE",
    "Kenneth BEDNAREK", "Akani SIMBINE",
    "Letsile TEBOGO", "Lamont Marcell JACOBS",
    "Benjamin RICHARDSON", "Christian COLEMAN"
]

# Dernier √©tat connu par athl√®te
df_latest = df_Xit.sort_values("Date").dropna(subset=["PB_so_far"]).groupby("Name").last().reset_index()

# === 3. Construction du DataFrame de simulation ===
df_sim = df_latest[df_latest["Name"].isin(athletes)].copy()
df_sim["Date"] = event_date
df_sim["Wind_float"] = np.random.uniform(-2, 2, size=len(df_sim))
df_sim["race_season_rank"] = 1
df_sim["RoundType"] = "Final"
df_sim["IsMajorEvent"] = True
df_sim["TimeGap"] = np.nan
df_sim["SB_so_far"] = np.nan
df_sim["avg_last_3"] = np.nan

# Calcul Age et HomeRace
df_sim["Age"] = df_sim.apply(lambda row: (event_date - pd.to_datetime(row["Date"])).days / 365.25 + row["Age"], axis=1)
df_sim["HomeRace"] = df_sim["Country"].apply(lambda nat: 1 if nat == "FRA" else 0)

# On garde les colonnes dans le bon ordre
df_sim = df_sim[features_finales if "Mark_clean" in df_sim.columns else [c for c in features_finales if c != "Mark_clean"]]

# === 4. Encodage et pr√©dictions ===
X_sim = df_sim.drop(columns=["Name", "Date"], errors="ignore")
X_sim_encoded = pd.get_dummies(X_sim)

# Aligne avec le training set (X_train) si certaines colonnes manquent ou sont en trop
X_sim_encoded = X_sim_encoded.reindex(columns=X_train.columns, fill_value=0)

# Pr√©dictions
df_sim["q05"] = model_q05.predict(X_sim_encoded)
df_sim["q50"] = model_q50.predict(X_sim_encoded)
df_sim["q95"] = model_q95.predict(X_sim_encoded)

# R√©sultat final pour simulation de duels
resultats_paris2025 = df_sim[["Name", "q05", "q50", "q95"]]

# D√©finition des duels
duels = [
    ("Kendal WILLIAMS", "Kishane THOMPSON"),
    ("Christian MILLER", "Ferdinand OMANYALA"),
    ("Emmanuel MATADI", "Noah LYLES"),
    ("Shainer REGINFO MONTOYA", "Fred KERLEY"),
    ("Ackeem BLAKE", "Oblique SEVILLE"),
    ("Kenneth BEDNAREK", "Akani SIMBINE"),
    ("Letsile TEBOGO", "Lamont Marcell JACOBS"),
    ("Benjamin RICHARDSON", "Christian COLEMAN")
]

# Indexation des temps pr√©vus
preds = resultats_paris2025.set_index("Name")

# Simulation des duels
results = []

for a1, a2 in duels:
    t1 = preds.loc[a1, "q50"]
    t2 = preds.loc[a2, "q50"]
    winner = a1 if t1 < t2 else a2 if t2 < t1 else np.random.choice([a1, a2])
    results.append({
        "Duel": f"{a1} vs {a2}",
        "Winner": winner,
        "q50_A": round(t1, 3),
        "q50_B": round(t2, 3)
    })

df_duels = pd.DataFrame(results)
print("üèÅ R√©sultats simul√©s ‚Äì The Sprinters' War, Paris 2025")
print(df_duels)
```

```python
# 1. Reconstruction des r√©sultats individuels
rows = []
for _, row in df_duels.iterrows():
    a1, a2 = row['Duel'].split(" vs ")
    q1, q2 = row['q50_A'], row['q50_B']
    w = row['Winner']
    rows.append({'Name': a1, 'q50': q1, 'Won': int(w == a1)})
    rows.append({'Name': a2, 'q50': q2, 'Won': int(w == a2)})

df_perfs = pd.DataFrame(rows)

# 2. Attribution des points de performance
df_perfs = df_perfs.sort_values("q50").reset_index(drop=True)
df_perfs["PerfPoints"] = 0
df_perfs.loc[0, "PerfPoints"] = 3
df_perfs.loc[1, "PerfPoints"] = 2
df_perfs.loc[2, "PerfPoints"] = 1

# 3. Points de victoire
df_perfs["WinPoints"] = df_perfs["Won"] * 3

# 4. Total
df_perfs["TotalPoints"] = df_perfs["WinPoints"] + df_perfs["PerfPoints"]

# 5. Classement final
df_ranking = df_perfs[["Name", "q50", "WinPoints", "PerfPoints", "TotalPoints"]].sort_values("TotalPoints", ascending=False).reset_index(drop=True)

# Affichage
print("üèÅ Classement final ‚Äì The Sprinters' War, Paris 2025")
print(df_ranking)
```

```python
# On regroupe les scores totaux pour chaque athl√®te
df_ranking_final = (
    df_ranking
    .groupby("Name")
    .agg({
        "q50": "min",              # meilleure perf (q50) sur les s√©ries
        "TotalPoints": "sum",      # total des points
        "q50": "count"             # nombre de courses (une ligne par course dans df_ranking)
    })
    .rename(columns={"q50": "NumRaces"})
    .sort_values("TotalPoints", ascending=False)
    .reset_index()
)

# R√©sultat final : classement cumul√©
print("üèÅ Classement final ‚Äì The Sprinters' War, Paris 2025")
print(df_ranking_final)
```

```python
# Fonction pour cr√©er un tableau Plotly
def create_plotly_table(df, title):
    fig = go.Figure(data=[go.Table(
        header=dict(
            values=list(df.columns),
            fill_color='#2a3f5f',
            font=dict(color='white', size=14),
            align='left'
        ),
        cells=dict(
            values=[df[col] for col in df.columns],
            fill_color=[['#f9f9f9', 'white'] * (len(df) // 2 + 1)],
            align='left',
            format=[".2f" if pd.api.types.is_numeric_dtype(df[col]) else None for col in df.columns]
        ))
    ])
    fig.update_layout(
        title=title,
        margin=dict(l=20, r=20, t=40, b=20),
        height=400 + len(df) * 20
    )
    return fig

# G√©n√©ration des figures
fig_top_perf = create_plotly_table(top_perf, "üèÅ Top 10 Individual Performances")
fig_sb = create_plotly_table(sb_athlete, "üî• Season Best per Athlete")
fig_avg = create_plotly_table(mean_perf_by_athlete, "üìä Average Performance per Athlete")

# Print des figures
fig_top_perf.show()
fig_sb.show()
fig_avg.show()

# Export HTML
fig_top_perf.write_html(f"../figures/top_10_individual_performances.html")
fig_sb.write_html(f"../figures/season_best_per_athlete.html")
fig_avg.write_html(f"../figures/average_perf_per_athlete.html")
```

```python
top8_names = ['Akani SIMBINE', 
              'Kishane THOMPSON', 
              'Fred KERLEY',  
              'Oblique SEVILLE', 
              'Christian COLEMAN',
              'Noah LYLES', 
              'Letsile TEBOGO', 
              'Lamont Marcell JACOBS']

# Filtrage des donn√©es top 8
df_filtered = all_results_season[all_results_season["Name"].isin(top8_names)].copy()

# S√©lection interactive sur la l√©gende
highlight = alt.selection_point(fields=['Name'], bind='legend')

# Courbes de performance (points + ligne)
lines = alt.Chart(df_filtered).mark_line(point=True).encode(
    x=alt.X('RaceOrder:O', title='Race Number'),
    y=alt.Y('q50:Q', title='Predicted Time (q50)', scale=alt.Scale(domain=[9.80, 10.30])),
    color='Name:N',
    opacity=alt.condition(highlight, alt.value(1), alt.value(0.1)),
    tooltip=['Name', 'RaceOrder', 'q50']
)

# Droites de r√©gression par sprinteur
regression = alt.Chart(df_filtered).transform_regression(
    'RaceOrder', 'q50', groupby=['Name']
).mark_line(strokeDash=[4, 4]).encode(
    x='RaceOrder:Q',
    y='q50:Q',
    color='Name:N',
    opacity=alt.condition(highlight, alt.value(1), alt.value(0.05))
)

# Combine les deux
chart = (lines + regression).add_params(highlight).properties(
    title='üèÅ Performance Trends + Season Regression ‚Äì Top 8 Sprinters',
    width=800,
    height=400
)

chart
```

```python
coefs = []

for name in top8_names:
    df_runner = all_results_season[all_results_season["Name"] == name]
    X = df_runner[["RaceOrder"]].values
    y = df_runner["q50"].values

    if len(X) >= 2:  # au moins deux points pour faire une r√©gression
        model = LinearRegression().fit(X, y)
        coefs.append({"Name": name, "Slope": model.coef_[0]})
    else:
        coefs.append({"Name": name, "Slope": None})

df_slope = pd.DataFrame(coefs).sort_values("Slope")

print(df_slope)
```

```python
chart_slope = alt.Chart(df_slope).mark_bar().encode(
    x=alt.X("Slope:Q", title="Regression Slope (Œî time per race)"),
    y=alt.Y("Name:N", sort="-x", title="Sprinter"),
    color=alt.condition("datum.Slope < 0", alt.value("#2ca02c"), alt.value("#d62728")),
    tooltip=["Name", "Slope"]
).properties(
    title="üìâ Sprinters' Season Progression (Negative = Improving)",
    width=600,
    height=300
)

chart_slope
```
